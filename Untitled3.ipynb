```python
!pip install -U --pre efficientnet
!pip install tensorflow

from efficientnet.tfkeras import EfficientNetB1
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense, BatchNormalization, Dropout, ReLU
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import seaborn as sns
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

# Constants
IMAGE_WIDTH = 224
IMAGE_HEIGHT = 224
IMAGE_SIZE = (IMAGE_WIDTH, IMAGE_HEIGHT)
IMAGE_CHANNELS = 3
BATCH_SIZE = 64
EPOCHS = 30

# Paths
train_path = "/content/drive/MyDrive/datapublik3/train"
test_path = "/content/drive/MyDrive/datapublik3/test"

# Load and preprocess data
def load_data(path):
    categories = []
    file_name = []
    for folder in os.listdir(path):
        folder_path = os.path.join(path, folder)
        if os.path.isdir(folder_path):
            for file in os.listdir(folder_path):
                file_path = os.path.join(folder_path, file)
                file_name.append(file_path)
                categories.append(folder)
    return pd.DataFrame({"image": file_name, "categories": categories}).sample(frac=1).reset_index(drop=True)

train_data = load_data(train_path)
test_data = load_data(test_path)

# Print unique categories to verify
print("Train data categories:", train_data['categories'].unique())
print("Test data categories:", test_data['categories'].unique())

# Display the count of samples in each category
print("Train data category distribution:")
print(train_data['categories'].value_counts())
print("\nTest data category distribution:")
print(test_data['categories'].value_counts())

# Plot category distribution
train_data['categories'].value_counts().plot.bar()
plt.show()

# Split data into training and validation sets
train_df, validate_df = train_test_split(train_data, test_size=0.10, random_state=20)
train_df = train_df.reset_index(drop=True)
validate_df = validate_df.reset_index(drop=True)

train_df['categories'].value_counts().plot.bar()
plt.show()

validate_df['categories'].value_counts().plot.bar()
plt.show()

# Data generators
train_datagen = ImageDataGenerator(
    rotation_range=30,
    rescale=1./255,
    shear_range=0.1,
    vertical_flip=True,
    zoom_range=0.2
)

validation_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_dataframe(
    dataframe=train_df,
    directory=None,
    x_col='image',
    y_col='categories',
    target_size=IMAGE_SIZE,
    class_mode='categorical',
    batch_size=BATCH_SIZE
)

validation_generator = validation_datagen.flow_from_dataframe(
    dataframe=validate_df,
    directory=None,
    x_col='image',
    y_col='categories',
    target_size=IMAGE_SIZE,
    class_mode='categorical',
    batch_size=BATCH_SIZE
)

# Model setup
base_model = EfficientNetB1(input_shape=(224, 224, 3), include_top=False, weights='noisy-student')

for layer in base_model.layers:
    layer.trainable = False

model = Sequential([
    base_model,
    Conv2D(640, kernel_size=(1, 1)),
    Flatten(),
    Dense(320, kernel_initializer='he_uniform'),
    BatchNormalization(),
    ReLU(),
    Dropout(0.2),
    Dense(2, activation='softmax')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# Model training
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=validation_generator
)

# Plot training history
plt.plot(history.history["accuracy"], label="Training Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.legend()
plt.show()

plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
plt.show()

# Evaluate model on test data
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_dataframe(
    dataframe=test_data,
    directory=None,
    x_col='image',
    y_col=None,
    class_mode=None,
    target_size=IMAGE_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

predictions = model.predict(test_generator, steps=np.ceil(len(test_data)/BATCH_SIZE))
predicted_classes = np.argmax(predictions, axis=1)

true_classes = test_data['categories'].replace(['BlackSkin', 'WhiteSkin'], [0, 1]).values

# Confusion matrix
cm = confusion_matrix(true_classes, predicted_classes)
cm_df = pd.DataFrame(cm, index=['BlackSkin', 'WhiteSkin'], columns=['BlackSkin', 'WhiteSkin'])

plt.figure(figsize=(8, 8))
sns.heatmap(cm_df, annot=True)
plt.title(f'CNN EfficientNet-B1 \nAccuracy: {accuracy_score(true_classes, predicted_classes):.3f}')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()

# Classification report
report = classification_report(true_classes, predicted_classes, target_names=['BlackSkin', 'WhiteSkin'])
print(report)

# Save model
model.save('geeksforgeeks.h5')
```
